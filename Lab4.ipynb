```python
ebola.iloc[0:10, 0:6]
```

```python
ebola_fill_backward = ebola.fillna(method='bfill')
print(f"Number of missing values after backward fill: {ebola_fill_backward.isnull().sum().sum()}")
print(ebola_fill_backward.iloc[0:10, 0:5])

'''
Explanation:
Backward fill (bfill) imputation replaces missing values with the value from the next valid observation in the same column. This means that if a cell has a missing value, it will be replaced with the value from the next row that has a non-missing value. This can be useful when the data has a temporal or sequential order and you believe the most recent valid value is a good approximation for the missing value. However, this approach can lead to biased estimates if the data is not ordered sequentially or if there are long stretches of missing data.
'''
```

9. Write a paragraph to highlight the differences between the fill foward/backward versus linear interpolation of the missing values.



```python
'''
Forward fill and backward fill methods are useful for handling missing values in time series data or data with a natural order. They essentially propagate the last known value forward or the next known value backward, respectively. These methods are simple and can be effective when the missing values are likely to be similar to the preceding or following values. However, they can introduce bias if the data has significant variations or trends.

Linear interpolation, on the other hand, estimates missing values by assuming a linear relationship between the surrounding data points. It draws a straight line between the known values and estimates the missing values based on this line. This method is more sophisticated than forward/backward fill and can be more accurate in cases where there is a gradual trend in the data. However, it may not be appropriate if the underlying relationship is non-linear.

In summary, forward/backward fill is suitable for data with a clear order and where missing values are expected to be close to the neighboring values. Linear interpolation is more suitable for data with a smooth, gradual trend, but it may not be accurate if the relationship between data points is non-linear or if there are sudden changes in the data.
'''
```
