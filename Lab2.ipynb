```python
17. Pivot the table further to hold information in the following format:
     	    class 	First 	Second 	Third
sex 	age
female 	(0, 18] 	0.909091 	1.000000 	0.511628
(18, 80] 	0.972973 	0.900000 	0.423729
male 	(0, 18] 	0.800000 	0.600000 	0.215686
(18, 80] 	0.375000 	0.071429 	0.133663
```

```python
# Define age bins and labels
age_bins = [0, 18, 80]
age_labels = ['(0, 18]', '(18, 80]']

# Create a new column for age groups
df_titanic['age_group'] = pd.cut(df_titanic['age'], bins=age_bins, labels=age_labels, right=True)

# Create a pivot table that groups by sex and age group, with class as columns
pivot_df = df_titanic.pivot_table(index=['sex', 'age_group'], columns='class', values='survived', aggfunc='mean')

pivot_df
```

```python
18.Pivot the table further to hold information in the following format:
fare 	survived
class 	First 	Second 	Third 	First 	Second 	Third
sex
female 	106.125798 	21.970121 	16.118810 	91.0 	70.0 	72.0
male 	67.226127 	19.741782 	12.661633 	45.0 	17.0 	47.0
```

```python
fare_pivot = df_titanic.pivot_table(index='sex', columns='class', values='fare', aggfunc='mean')

survived_pivot = df_titanic.pivot_table(index='sex', columns='class', values='survived', aggfunc='sum')

final_pivot = pd.concat([fare_pivot, survived_pivot], axis=1, keys=['fare', 'survived'])

final_pivot
```

```python
23. Write a query to print the name of employee along with the complete address of their department.
```

```python
sql_query = '''
SELECT employees.first_name, employees.last_name, locations.street_address,
locations.postal_code, locations.city, locations.state_province, locations.country_id
FROM employees
LEFT JOIN departments ON employees.department_id = departments.department_id
LEFT JOIN locations ON departments.location_id = locations.location_id
'''

pd.read_sql(sql_query, mydb)
```
